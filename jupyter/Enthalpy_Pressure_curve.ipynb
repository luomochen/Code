```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
workspace = '/mnt/d/Workspace/Data/SiO2/ptp/'
```

```python
def convert_ndarray(data, column_name):
    data.columns = data.columns.str.strip()
    numeric = pd.to_numeric(data[column_name])
    ndarray = np.array(numeric.tolist())
    return ndarray
```

```python
plt.rcParams.update({'font.size': 14})
```

```python
plt.figure(figsize=(8, 6))
plt.xlabel('Pressure (GPa)')
plt.ylabel('Enthalpy (eV)')
labels=['stishovite', 'coesite', 'quartz', '$PbO_2$-type', 'pyrite']
phase_list = ['stishovite', 'coesite', 'quartz', 'PbO2', 'pyrite']
chemical_unit_number = [2, 16, 3, 4, 4]
color_list = ['#e2041b', '#006e54', '#895b8a', '#b98c46', '#8f2e14', '#e6b422']
data_range = np.arange(0, 301, 25)
for i in range(len(phase_list)):   
    data = pd.read_csv(workspace+phase_list[i]+"/outcome.csv")
    pressure = convert_ndarray(data, 'pressure')
    enthalpy = convert_ndarray(data, 'enthalpy')
    enthalpy = enthalpy / chemical_unit_number[i]
    indices = np.where(np.isin(pressure, data_range))[0]
    if i == 0:
        ref_phase_enthalpy = enthalpy[indices]
    delta_enthalpy = enthalpy[indices] - ref_phase_enthalpy
    plt.plot(pressure[indices], delta_enthalpy, marker='o', 
             color=color_list[i], markersize=8,
             alpha=0.85, label=labels[i])
plt.legend()
```
